## Sparse matrix

Make a (random, directed) graph:

```raku
#% html 
#my $g = Graph::Circulant.new(12, 3);
my $g = Graph::Random.new(12, 13):d;
$g .= vertex-replace( ((^12).Str Z=> random-string(12, :3chars).sort).Hash);
$g.dot(engine => 'neato'):svg;
```

Make the sparse matrix from graph's edges:

```raku
#% html
my $smat = Math::SparseMatrix.new(edge-dataset => $g.edges(:dataset)):directed
```

Plot the corresponding dense matrix:

```raku
#% html
dot-matrix-plot($smat.dense-matrix, :4size, row-names => $smat.row-names, column-names => $smat.column-names):svg
```
